**Task:**

Write a function `print_word_lengths(words)` that takes a list of strings and prints each word and its length on its own line in the format:

`word: length`

Use an element-based `for` loop. The function does not return anything.

**Sample Output:**
```

```



`````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def print_word_lengths(words):
  ''' takes a list of strings and prints each word and its length on its own line'''

  # for loop to iterate through each string in the list
  for w in range (len(words)):
    
    # print the word in each index, then write the length of that word in each index
    print(f'{words[w]}: {len(words[w])}')
`````

**Review Problem 2**

**Concept:** *Processing a list with a definite loop, using accumulator pattern.*

**Task:**

Write a function `total_length(words)` that returns the sum of the lengths of all words in a list using the accumulator pattern.

Write a second version using list comprehension.

**Sample Output:**

```
total_length(['data', 'science', 'rocks'])
```
should return `16`




`````python
def total_length(words):
  ''' returns the sum of the lengths of all words in a list'''

  # defining an accumulator
  sum_words = 0

  # for loop to iterate between through each string in the list
  for w in range (len(words)):
  
    # gets the length of each word, then adds it to the accumulator
    sum_words = sum_words + len(words[w])

  # gives the total
  return sum_words

# # version using list comprehension

def total_length_two(words):
  '''returns the sum of the lengths of all words in a list'''

  # after getting the length of each word in the list, get the sum of the lengths
  return sum([len(w) for w in words])

`````

**Problem 5**

**Concept:** *Process a string, find unique characters.*

**Task:**

Write a function `unique_letters(s)` that returns a list of unique characters in a string using:

* list comprehension

* set

**Sample Output:**
```
print(unique_letters('hello world'))
```
should return
```
unique letters: ['h', 'e', 'l', 'o', ' ', 'w', 'r', 'd']
```



`````python
def unique_letters(s):
  '''returns a list of unique characters in a string'''

  # created a new variable that will hold the new list of unique characters
  new_word_set = set()
  
  # go through each character in s. If we haven't seen it before, keep it. If we have, skip it. Then add it to the set. return the result in the new list
  return [ch for ch in s if ch not in new_word_set and not new_word_set.add(ch)]
 

`````

should print






`````python
def print_values(d):
  '''prints each key or value pair in a dictionary on its own line'''

  # a loop for the key and the value
  for k, v in d.items():

    # print both the key and value on their own line
    print(f'{k}: {v}')
`````





`````python
def word_score(word):
  '''returns the Scrabble score of a word using the dictionary scrabble_scores'''

  # create accumulator variable
  score = 0

  # for loop to go through each character in the word
  for ch in word:
    
    # go through the dictionary, and find each character's key. Then add it to the accumulator
    score = score + scrabble_scores[ch]

  # return the word and the sum of the characters
  return (f'{word}: {score}')
      
 
`````

**Problem 9**

**Concept:** *Using list data type, dictionary data type, loop, dictionary element access, accumulator pattern.*

**Task:**

Write a function `total_points(games_results, team_name)` that returns the total points scored by a team across all games.

**Sample Output:**
```
games_results = [
    {'A': 5, 'B': 7},
    {'A': 10, 'C': 3}
]
print(total_points(games_results, 'A'))
```
should return
```
total points for A: 15
```



`````python
games_results = [
  {'A': 5, 'B': 7}, 
  {'A': 10, 'C': 3}
]

def total_points(games_results, team_name):
  '''returns the total points scored by a team across all games'''

 # create accumulator variable
  score = 0
  
  # loop to look at each game in the game results dictionary
  for game in games_results:

    # if the team name entered is in a game
    if team_name in game:

      # take the value of the key and add it to the accumulator
      score = score + game[team_name]
    
  # return the total score
  return score
`````

`````python
def sort_games_by_score(games, team_name):
  '''returns a new list sorted by a team's score in descending order'''

  # sort the games with the lambda key, getting the values of the team name
  # keys in the game dictionary, and doing it descending.
  games.sort(key = lambda game: game[team_name], reverse=True)

  # return the dictionary descending
  return games
  
  


`````

**Problem 11**

**Concept:** *Using dictionary data type, loop, decision statements.*

**Task:**

Write a function `best_score(games, team_name)` that returns the game dictionary where the team scored the most points.

**Sample Output:**
```
games = [
    {'A': 8, 'B': 10},
    {'A': 15, 'C': 7},
    {'A': 12, 'D': 9}
]
print(best_score(games, 'A'))
```
should return
```
{'A': 15, 'C': 7}
```




`````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def best_score(games, team_name):
  '''returns the game dictionary where the team scored the most points'''
  
  # an accumulator variable to hold the best game score
  highest_score = 0

  # an accumulator variable that's empty to put the best game in
  best_game = {}

  # loop through each game dictionary in the dictionary
  for game in games:

    # if the value of the team name key has the highest score
    if game[team_name] > highest_score:

      # make that key the new highest score
      highest_score = game[team_name]

      # for the highest scoring game, place that game into the empty variable
      best_game = game
  
  # return the game with the highest score
  return best_game
      
`````

**Problem 12**

**Concept:** *Using list data type, dictionary data type, sorting.*

**Task:**

Write a function `best_word(words)` that returns the word with the highest Scrabble score.

**Sample Output:**
```
best_word(['dog', 'quiz', 'cat'])
```
should return `quiz`, because `quiz` has the highest scrabble score.




`````python
def best_word(words):
  '''returns the word with the highest Scrabble score'''

  # created accumulator for the highest score of the scrabble word
  highest_score = 0

  # empty list to store the best word 
  best_scrabble = []

  # going through each word in the words list that is entered into the function
  for word in words:

    # created accumulator to get a score of each of the words
    new_score = 0

    # go through each letter in each word and get the score of the letters
    for letters in word:

      # add the scrabble word scores up
      new_score = new_score + scrabble_scores[letters]

    # if the scrabble score of the word is larger than the highest score
    if new_score > highest_score:

      # replace the highest score with the new score from the word
      highest_score = new_score

      # place the word with the highest score in the empty list
      best_scrabble = word

  # return the best word
  return best_scrabble
`````
